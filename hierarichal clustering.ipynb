```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import DBSCAN

# Step 1: Create a Synthetic Dataset
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Convert to DataFrame for easier handling
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])

# Step 2: Perform Hierarchical Clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

Z = linkage(X_scaled, method='ward')

plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()

# Step 3: Perform DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)

# Add the cluster labels to the dataframe
df['Cluster'] = clusters

# Visualize the clusters
sns.pairplot(df, hue='Cluster', palette='viridis')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage

# Membuat Dataset Sintetis
X, y = make_blobs(n_samples=50, centers=3, random_state=42)

# Mengonversi ke DataFrame untuk memudahkan pengelolaan
df = pd.DataFrame(X, columns=['Fitur1', 'Fitur2'])

# Melakukan standarisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Melakukan Hierarchical Clustering
Z = linkage(X_scaled, method='ward')

# Membuat dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z, truncate_mode='lastp', p=12, leaf_rotation=90., leaf_font_size=12., show_contracted=True)
plt.title('Dendrogram untuk Hierarchical Clustering')
plt.xlabel('Indeks Sampel atau (Cluster Size)')
plt.ylabel('Jarak')
plt.show()
```

```python
import numpy as np

# Data set contoh
X = np.array([[1, 2], [5, 8], [1.5, 1.8], [8, 8], [1, 0.6], [9, 11]])
from sklearn.cluster import KMeans

# Inisialisasi model K-Means dengan jumlah cluster yang diinginkan
kmeans = KMeans(n_clusters=2)

# Melatih model pada data
kmeans.fit(X)

# Mendapatkan label dari cluster yang diberikan oleh model
labels = kmeans.labels_

# Mendapatkan pusat dari cluster yang diberikan oleh model
centers = kmeans.cluster_centers_

print("Label cluster:", labels)
print("Pusat cluster:", centers)
```

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Data set contoh
X = np.array([[1, 2], [5, 8], [1.5, 1.8], [8, 8], [1, 0.6], [9, 11]])

# Menggunakan metode linkage single dan Euclidean distance
Z = linkage(X, method='single', metric='euclidean')

# Menampilkan dendrogram
plt.figure(figsize=(10, 5))
plt.title('Dendrogram Hierarchical Clustering')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
dendrogram(Z)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Data set contoh
X = np.array([[1, 2], [5, 8], [1.5, 1.8], [8, 8], [1, 0.6], [9, 11]])

# Menginisialisasi model K-Means dengan jumlah cluster yang diinginkan
kmeans = KMeans(n_clusters=2)

# Melatih model pada data
kmeans.fit(X)

# Mendapatkan label dari cluster yang diberikan oleh model
labels = kmeans.labels_

# Mendapatkan pusat dari cluster yang diberikan oleh model
centers = kmeans.cluster_centers_

# Menampilkan hasil klastering
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='x')
plt.title('K-Means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Memuat dataset iris
iris = load_iris()
X = iris.data  # Fitur-fitur dari dataset
y = iris.target  # Label dari dataset

# Menginisialisasi model K-Means dengan jumlah cluster yang diinginkan
kmeans = KMeans(n_clusters=3)

# Melatih model pada data
kmeans.fit(X)

# Mendapatkan label dari cluster yang diberikan oleh model
labels = kmeans.labels_

# Menampilkan hasil klastering
plt.figure(figsize=(10, 5))

# Plot untuk fitur panjang kelopak dan panjang mahkota
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 2], c=labels, cmap='viridis', s=50, alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 2], c='red', s=200, marker='x')
plt.xlabel('Panjang Kelopak (cm)')
plt.ylabel('Panjang Mahkota (cm)')
plt.title('K-Means Clustering (Sepal Length vs Petal Length)')

# Plot untuk fitur lebar kelopak dan lebar mahkota
plt.subplot(1, 2, 2)
plt.scatter(X[:, 1], X[:, 3], c=labels, cmap='viridis', s=50, alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 3], c='red', s=200, marker='x')
plt.xlabel('Lebar Kelopak (cm)')
plt.ylabel('Lebar Mahkota (cm)')
plt.title('K-Means Clustering (Sepal Width vs Petal Width)')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Memuat dataset penguins
penguins = sns.load_dataset("penguins")

# Menghapus baris dengan nilai yang hilang
penguins = penguins.dropna()

# Memilih fitur yang akan digunakan untuk klastering
X = penguins[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]

# Menginisialisasi model K-Means dengan jumlah cluster yang diinginkan
kmeans = KMeans(n_clusters=3)

# Melatih model pada data
kmeans.fit(X)

# Mendapatkan label dari cluster yang diberikan oleh model
labels = kmeans.labels_

# Menambahkan label klaster ke dalam dataset
penguins['cluster'] = labels

# Plot hasil klastering
sns.pairplot(penguins, hue='cluster', palette='viridis')
plt.suptitle('K-Means Clustering on Penguins Dataset')
plt.show()
```
